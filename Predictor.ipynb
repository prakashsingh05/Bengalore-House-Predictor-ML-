```python
def remove_outliers_sqft(df):
    df_output = pd.DataFrame()
    for key, subdf in df.groupby('location'):
        m = np.mean(subdf.price_per_sqft)
        st = np.std(subdf.price_per_sqft)
        
        # Keep only values within (mean - std) and (mean + std)
        gen_output = subdf[(subdf.price_per_sqft > (m - st)) & 
                           (subdf.price_per_sqft <= (m + st))]
        
        df_output = pd.concat([df_output, gen_output], ignore_index=True)
    
    return df_output
df = remove_outliers_sqft(df)
```

```python
df.describe()
```

```python
def bhk_outlier_remover(df):
    exclude_indices = np.array([])   # store indices of outlier rows
    
    # Loop over each location
    for location, location_df in df.groupby('location'):
        bhk_stats = {}
        
        # Step 1: Calculate mean, std, count for each BHK in this location
        for bhk, bhk_df in location_df.groupby('bhk'):
            bhk_stats[bhk] = {
                'mean': np.mean(bhk_df.price_per_sqft),
                'std': np.std(bhk_df.price_per_sqft),
                'count': bhk_df.shape[0]
            }
        
        # Step 2: Compare each BHK group with (bhk-1) group
        for bhk, bhk_df in location_df.groupby('bhk'):
            stats = bhk_stats.get(bhk - 1)
            
            # Only if (bhk-1) group exists and has enough samples
            if stats and stats['count'] > 5:
                # Find rows where higher BHK price_per_sqft < lower BHK mean
                exclude_indices = np.append(
                    exclude_indices,
                    bhk_df[bhk_df.price_per_sqft < stats['mean']].index.values
                )
    
    # Remove outliers
    return df.drop(exclude_indices, axis='index')
```

```python
df = bhk_outlier_remover(df)
```

```python
X = df.drop(columns=['price'])
y = df['price']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression, Lasso, Ridge
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline
from sklearn.metrics import r2_score
```

```python
X_train,X_test, y_train,y_test = train_test_split(X,y,test_size =0.2, random_state=0)
```

```python
print(X_train.shape)
print(X_test.shape)
```

Applying Linear Regression

```python
column_trans = make_column_transformer(
    (OneHotEncoder(sparse_output=False, handle_unknown='ignore'), ['location']),
    remainder='passthrough'
)

scaler = StandardScaler()
lr = LinearRegression()

# Pipeline
pipe = make_pipeline(column_trans, scaler, lr)

# Train
pipe.fit(X_train, y_train)

# Predict
y_pred_lr = pipe.predict(X_test)

# Evaluate
print("R2 Score:", r2_score(y_test, y_pred_lr))
```

Applying Lasso

```python
lasso = Lasso()
```

```python
pipe  =make_pipeline(column_trans,scaler, lasso)
```

```python
pipe.fit(X_train,y_train)
```

```python
y_pred_lasso = pipe.predict(X_test)
r2_score(y_test, y_pred_lasso)
```

Applying Ridge

```python
ridge = Ridge()
```

```python
pipe = make_pipeline(column_trans,scaler,ridge)
```

```python
pipe.fit(X_train,y_train)
```

```python
y_pred_ridge = pipe.predict(X_test)
r2_score(y_test, y_pred_ridge)
```

```python
print("No Regularization (LinearRegression):", r2_score(y_test, y_pred_lr))
print("Lasso:", r2_score(y_test, y_pred_lasso))
print("Ridge:", r2_score(y_test, y_pred_ridge))
```
